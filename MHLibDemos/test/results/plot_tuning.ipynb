```python
import pandas as pd
import re

df = pd.read_csv("construction_results_experiments1.csv")

# --------- Helper extractors ---------
def extract_instance_number(name):
    m = re.search(r'instance(\d+)', name)
    return int(m.group(1)) if m else 999999

def extract_nreq(name):
    m = re.search(r'nreq(\d+)', name)
    return int(m.group(1)) if m else 999999

# --------- Group the results ---------
summary = df.groupby(["instance", "alg", "alpha"]).agg({
    "obj": ["mean", "std", "min"],
    "total_time": "mean",
    "fairness": "mean"
}).reset_index()

# --------- Extract sorting keys ---------
summary["nreq"] = summary["instance"].apply(extract_nreq)
summary["inst_num"] = summary["instance"].apply(extract_instance_number)

# --------- Sort: first nreq=50, then nreq=200 ---------
summary_sorted = summary.sort_values(
    ["nreq", "inst_num", "alg", "alpha"],
    ascending=[True, True, True, True]
)

# Remove helper columns
summary_sorted = summary_sorted.drop(columns=["nreq", "inst_num"])

# --------- Optional rounding ---------
summary_sorted[("obj", "mean")] = summary_sorted[("obj", "mean")].round(2)
summary_sorted[("obj", "std")]  = summary_sorted[("obj", "std")].round(2)
summary_sorted[("obj", "min")]  = summary_sorted[("obj", "min")].round(2)
summary_sorted[("total_time", "mean")] = summary_sorted[("total_time", "mean")].round(2)
summary_sorted[("fairness", "mean")]   = summary_sorted[("fairness", "mean")].round(3)

# --------- Convert to LaTeX ---------
latex_table = summary_sorted.to_latex(index=False, escape=False)
print(latex_table)
```

```python
import pandas as pd

df = pd.read_csv("grasp_tuning.csv")

# extract instance number
df["inst_num"] = df["instance"].str.extract(r"instance(\d+)_")[0].astype(int)

# extract nreq (50 or 200) from filename
df["nreq"] = df["instance"].str.extract(r"nreq(\d+)_")[0].astype(int)

# Sort by nreq first (50 before 200), then inst_num, then niters, then seed
df_sorted = df.sort_values(["nreq", "inst_num", "niters", "seed"]).reset_index(drop=True)

# Select and reorder columns
df_sorted = df_sorted[[
    "instance", "alpha", "niters", "seed",
    "obj", "total_time", "fairness", "runtime"
]]

# Round numerical values
df_rounded = df_sorted.copy()
df_rounded["obj"] = df_rounded["obj"].round(2)
df_rounded["total_time"] = df_rounded["total_time"].round(2)
df_rounded["fairness"] = df_rounded["fairness"].round(3)
df_rounded["runtime"] = df_rounded["runtime"].round(3)

# Convert to LaTeX
latex_table = df_rounded.to_latex(
    index=False,
    escape=False,
    column_format="lrrrrrrr",
    longtable=True
)

print(latex_table)
```
